```python
#APLICAMOS IQR PARA LOS DATOS ATIPICOS DE LA VARIABLE INCOME:

#Primero calculamos los cuartiles Q1 y Q3
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

#Luego calculamos el rango intercuartílico (IQR)
IQR = Q3 - Q1

#Ahora definimos los límites inferior y superior
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

#Finalmente reemplazamos los valores atípicos con la mediana de la variable
df.loc[df['Income'] < limite_inferior, 'Income'] = df['Income'].median()
df.loc[df['Income'] > limite_superior, 'Income'] = df['Income'].median()

#Visualizamos ahora como queda la distribución:

plt.figure(figsize=(8, 4))
sns.boxplot(x='Income', data=df)
plt.title('Distribución de la variable "Income"')
plt.xlabel('Income')
plt.grid(True)
plt.show()
```

# 4.4.1 VARIABLES CATEGORICAS

PARA LAS SIGUIENTES VARIABLES CATEGÓRICAS SE APLICARÁ LA PRUEBA ESTADÍSTICA CHI-CUADRADO, UTILIZANDO EL VALOR P-VALUE A FIN DE PODER ACEPTAR O RECHAZAR CADA UNA DE LAS HIPÓTESIS NULAS PLANTEADAS.

***1º Hipótesis Education:***
- Hipótesis nula: Los clientes con grado académico (Graduate) presentan igual aceptación a la oferta en la última campaña de marketing que los clientes que no tienen un grado académico (Non-graduate).
- Hipótesis alternativa:  Los clientes con grado académico (Graduate) presentan diferente aceptación a la oferta en la última campaña de marketing que los clientes que no tienen un grado académico (Non-graduate).

```python
#Graficamos de barras apiladas para visualizar la proporción de respuestas positivas y negativas de los clientes en función de su nivel educativo.
#Para ello creamos una tabla cruzada entre 'Education' y 'Response' con su respectiva proporción respecto al total.
table = pd.crosstab(df['Response'], df['Education'])
table1= table.apply(lambda r: r/r.sum())
print(table1)
```

```python
#Asignamos los valores de la tabla en listas a fin de ingresarlos al gráfico de barras
df_Education=['Graduate','Non_Graduate']
df_Response_No=table1.iloc[0].tolist()
df_Response_Si=table1.iloc[1].tolist()
print(df_Education)
print(df_Response_No)
print(df_Response_Si)
```

```python
#Gráfico de barras apiladas:
fig, ax = plt.subplots()
ax.bar(df_Education, df_Response_No)
ax.bar(df_Education, df_Response_Si, bottom = df_Response_No)
ax.legend(title='Response', labels=['No', 'Sí'], loc = 'upper right')
ax.set_xlabel('Education')
ax.set_ylabel('Proporción')
plt.show()
```

```python
#Realizamos ahora la prueba de chi-cuadrado:
chi2, p_val, _, _ = chi2_contingency(table)

# Mostrar los resultados
print("Chi-squared Test")
print(f"Chi-squared = {chi2:.4f}, p-value = {p_val:.4f}")
```

De acuerdo con los resultados, se rechaza la Hipótesis nula y se acepta la Hipótesis alternativa que nos dice que la aceptación de la oferta en la última campaña de marketing de los clientes con un grado académido es diferente a la de los clientes sin grado académico.

***2º Hipótesis Marital_Status:***
- Hipótesis nula: Los clientes solteros (Single) tienen igual aceptación de la oferta en la última campaña de marketing que los clientes casados (Married).
- Hipótesis alternativa: Los clientes solteros (Single) tienen diferente aceptación de la oferta en la última campaña de marketing que los clientes casados (Married).

```python
#Graficamos de barras apiladas para visualizar la proporción de respuestas positivas y negativas de los clientes en función de su estado civil.
#Para ello creamos una tabla cruzada entre 'Marital_Status' y 'Response' con su respectiva proporción respecto al total.
table = pd.crosstab(df['Response'], df['Marital_Status'])
table1= table.apply(lambda r: r/r.sum())
print(table1)

df_Marital_Status=['Married','Single']
df_Response_No=table1.iloc[0].tolist()
df_Response_Si=table1.iloc[1].tolist()
print(df_Marital_Status)
print(df_Response_No)
print(df_Response_Si)

fig, ax = plt.subplots()
ax.bar(df_Marital_Status, df_Response_No)
ax.bar(df_Marital_Status, df_Response_Si, bottom = df_Response_No)
for bar in ax.patches:
  ax.text(bar.get_x() + bar.get_width() / 2,
          bar.get_height() / 2 + bar.get_y(),
          round(bar.get_height(), 2), ha = 'center',
          color = 'w', weight = 'bold', size = 10)
ax.legend(title='Response', labels=['No', 'Sí'], loc = 'upper right')
ax.set_xlabel('df_Marital_Status')
ax.set_ylabel('Proporción')
plt.show()

#Realizamos ahora la prueba de chi-cuadrado:
chi2, p_val, _, _ = chi2_contingency(table)

# Mostrar los resultados
print("Chi-squared Test")
print(f"Chi-squared = {chi2:.4f}, p-value = {p_val:.4f}")
```

PARA LAS SIGUIENTES VARIABLES NUMÉRICAS SE APLICARÁ LA PRUEBA ESTADÍSTICA T-STUDENT, UTILIZANDO EL VALOR P-VALUE A FIN DE PODER ACEPTAR O RECHAZAR CADA UNA DE LAS HIPÓTESIS NULAS PLANTEADAS.

***3º Hipótesis Income:***
- Hipótesis nula: Los clientes con ingresos menores e iguales a la media rechazan la oferta en la última campaña de marketing.
- Hipótesis alternativa:  Los clientes con ingresos mayores a la media aceptan la oferta en la última campaña de marketing.

```python
#Boxplot para visualizar la distribución de los datos de la variable Income en función de las respuestas de los clientes.
plt.figure(figsize=(8, 4))
sns.boxplot(y = 'Income', hue = 'Response', data=df)
plt.legend(title='Response', labels=['No', 'Sí'], loc = 'upper right')
plt.ylabel('Income')
plt.show()

#Calculamos la media de los Ingresos de los clientes:
mean_income = df['Income'].mean()

#Luego filtramos en una muestra a los clientes con ingresos menores e iguales a la media y clientes con ingresos mayores a la media:
muestra = df[df['Income'] <= mean_income]

#Aplicamos  la prueba t-student para comparar las medias de la muestra y la población.
t_stat, p_val = ttest_ind(muestra['Response'], df['Response'])

#Imprimimos el pvalue
print("T-Student Test")
print("El valor de la media es: ", mean_income)
print("El estadísticio t es:", t_stat)
print("El pvalue es:", p_val)
```

De acuerdo con los resultados, p-value = 0.006 es menor que 0.05 por lo que se rechaza la Hipótesis nula y no se rechaza la Hipótesis alternativa que nos dice que los clientes con ingresos mayores a la media aceptan la oferta en la última campaña de marketing.

```python
#Ahora seleccionamos aquellas que ejercen un efecto significativo sobre el target

#Var. Categóricas (con prueba de Chi Cuadrado)

#Creamos una lista para almacenar las variables categóricas significativas
sig_cat_var = []

for i in cat_vars:
    table = pd.crosstab(df[i], df[target])
    chi2_stat, p_val, dof, expected = chi2_contingency(table)
    print(f"{i}, Chi2 Stat: {chi2_stat}, p-value: {p_val}")
    if p_val < 0.05:
        sig_cat_var.append(i)

#Ahora filtramos aquellas que tienen pvalue < 0.05, esto significa que tienen un efecto significativo, se rechaza la hipótesis nula de que las medias son iguales.
print("Variables categóricas significativas:")
print(sig_cat_var)
```

```python
#Var. Numéricas (con prueba de T_Student)

#Creamos una lista para almacenar las variables categóricas significativas
sig_num_var = []

for i in num_vars:
    table = pd.crosstab(df[i], df[target])
    t_stat, p_val = ttest_ind(df[i], df[target])
    print(f"{i}, t Stat: {t_stat}, p-value: {p_val}")
    if p_val < 0.05:
        sig_num_var.append(i)

#Ahora filtramos aquellas que tienen pvalue < 0.05, esto significa que tienen un efecto significativo, se rechaza la hipótesis nula de que las medias son iguales.
print("Variables numéricas significativas:")
print(sig_num_var)
```
